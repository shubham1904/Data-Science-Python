```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('stopwords')
```

```python
ps = PorterStemmer()

sup.review = sup.review.str.lower().str.replace('[^a-z]', ' ').str.split()
sup.review = sup.review.apply(lambda x : ' '.join([ps.stem(word) for word in x if word not in set(stopwords.words('english'))]))

sup.head()
```

```python
from sklearn.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
nltk.download('wordnet')
sup.review = sup.review.apply(lambda x : ' '.join([wnl.lemmatize(word) for word in x]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X  = cv.fit_transform(sup.review).toarray()
pd.DataFrame(X, columns=cv.get_feature_names())
```
